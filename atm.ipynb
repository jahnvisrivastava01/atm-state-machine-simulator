```python
class ATM:

    def __init__(self):
        self.state = "Idle"
        self.balance = 5000
        self.pin = "1234"

    def insert_card(self):
        if self.state == "Idle":
            self.state = "Card Inserted"
            print("Card inserted successfully.")
        else:
            print("Card is already inserted.")

    def enter_pin(self, entered_pin):
        if self.state == "Card Inserted":
            if entered_pin == self.pin:
                self.state = "PIN Verified"
                print("✅ PIN verified successfully.")
            else:
                print("❌ Incorrect PIN.")
        else:
            print("⚠️ Please insert the card first.")

    def withdraw(self, amount):
        if self.state == "PIN Verified":
            if amount <= self.balance:
                self.balance -= amount
                print(f"✅ ₹{amount} withdrawn successfully.")
                print(f"💰 Remaining Balance: ₹{self.balance}")
            else:
                print("❌ Insufficient balance.")
        else:
            print("⚠️ Verify PIN first.")

    def deposit(self, amount):
        if self.state == "PIN Verified":
            self.balance += amount
            print(f"✅ ₹{amount} deposited successfully.")
            print(f"💰 Current Balance: ₹{self.balance}")
        else:
            print("⚠️ Verify PIN first.")

    def check_balance(self):
        print(f"💰 Current Balance: ₹{self.balance}")

    def exit(self):
        self.state = "Idle"
        print("✅ Transaction completed.")
        print("💳 Card removed.")
        print("🙏 Thank you for using our ATM!")
```

```python
atm = ATM()
```

```python
atm.insert_card()
```

```python
print("Current State:", atm.state)
```

```python
atm.enter_pin("1234")
```

```python
print("Current State:", atm.state)
```

```python
atm.withdraw(1000)
```

```python
atm.deposit(500)
```

```python
atm.check_balance()
```

```python
atm.exit()
```
